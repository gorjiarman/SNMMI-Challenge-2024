```python
df = pd.read_excel("/content/drive/MyDrive/Challenge_SNMMI_2023/Data/SNMMI_CHALLENGE_TRAINING_V22OCT2023_Update Oct 26.xlsx", index_col=False)
df.fillna(df.mean(), inplace=True)
df = df.iloc[:,:]

tdf = pd.read_excel("/content/drive/MyDrive/Challenge_SNMMI_2023/Data/SNMMI_CHALLENGE_TESTING_V01112023.xlsx", index_col=0)
tdf.fillna(tdf.mean(), inplace=True)
tdf = tdf.iloc[:,:]
tdata_x = tdf

df['Event'] = df['Event'].astype('bool')
evat = df[['Event','Outcome_PFS']].to_records(index=False)
data_x = df.iloc[:,3:]
data_y = evat
data_y
```

#Defs

```python
def score_survival_model(model, X, y):
    prediction = model.predict(X)
    result = concordance_index_censored(y['Event'], y['Outcome_PFS'], prediction)
    return result[0]
warnings.simplefilter('ignore', ConvergenceWarning)
def append_row(df, row):
    return pd.concat([
                df,
                pd.DataFrame([row], columns=row.index)]
           ).reset_index(drop=True)
```

#Componentwise Gradient Boosting Survival Analysis


```python
cwgb = ComponentwiseGradientBoostingSurvivalAnalysis(random_state=42)
parameters = {
              'model__loss': ['coxph' , 'squared' ,'ipcwls' ],
              'model__learning_rate': [1, 10, 100, 500],
              'model__dropout_rate': [0.1, 0.5, 0.9],
              'model__n_estimators': [10, 50, 200],
            }

cv = KFold(n_splits=5, shuffle=True, random_state=0)

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn import linear_model
#pca
pca = PCA(n_components=2)

print(data_x.shape)


rfe = RFE(estimator=AdaBoostClassifier(), n_features_to_select=50)
data_x_rfe = rfe.fit_transform(data_x,data_y)

print(data_x_rfe.shape)

randm_src = RandomizedSearchCV(
    Pipeline(steps = [('scaler',MinMaxScaler()),("model", cwgb)]),
    param_distributions = parameters,
    cv=cv,
    verbose=5,
    scoring = score_survival_model,
    n_iter = 50,
    n_jobs=1).fit(data_x_rfe, data_y)



cwgb_pred =  Pipeline(steps = [('scaler', MinMaxScaler()),("model", cwgb)])
cwgb_pred.set_params(**randm_src.best_params_)
cwgb_pred.fit(data_x, data_y)

cv_results = cross_validate(cwgb_pred, data_x, data_y, cv=5,return_train_score=True)

new_row = ['CWGB','all',randm_src.best_params_,cv_results['test_score'].mean(),cv_results['train_score'].mean()]

with open('/content/drive/MyDrive/Challenge_SNMMI_2023/Results/Amirali & Nima/Survival Last/Survival.csv','a') as fd:

    writer_object = writer(fd)
    writer_object.writerow(new_row)
```

#Predict results and saving it

```python
df = pd.DataFrame(columns=['pat_no','year1','year2','year3'])

ooprobabs = cwgb_pred.predict_survival_function(tdata_x,return_array=True).tolist()
warnings.simplefilter('ignore', ConvergenceWarning)

for a in range(len(ooprobabs)):
  ooprobabs=cwgb_pred.predict_survival_function(tdata_x)[a]
  times = ooprobabs.x.tolist()

  probabs=cwgb_pred.predict_survival_function(tdata_x,return_array=True)[a]
  dfprobabs = probabs.tolist()

  # Create an interpolation function using linear interpolation
  interpolation_func = interpolate.interp1d(times, dfprobabs, kind='linear')

  # Use the interpolation function to estimate the survival probability at the desired time
  year1_estimated_survival_prob = interpolation_func(12)
  year2_estimated_survival_prob = interpolation_func(24)
  year3_estimated_survival_prob = interpolation_func(36)

  # print('pat_no : '+str(a+1)+' First year : '+str(year1_estimated_survival_prob)+' Second year : '+str(year2_estimated_survival_prob)+' Third year : '+str(year3_estimated_survival_prob))
  new_row = pd.Series({'pat_no':str(a+1),'year1': str(year1_estimated_survival_prob),'year2':str(year2_estimated_survival_prob),'year3':str(year3_estimated_survival_prob)})
  df = append_row(df, new_row)
df.to_excel('/content/drive/MyDrive/Challenge_SNMMI_2023/Results/Arman/sr_cwgb.xlsx')
df
```
